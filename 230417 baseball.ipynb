```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 예시 데이터 생성
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # 독립 변수 X
y = np.array([2, 4, 6, 8, 10])               # 종속 변수 y

# 선형 회귀 모델 학습
model = LinearRegression()  # 선형 회귀 모델 객체 생성
model.fit(X, y)             # 모델 학습

# 학습된 모델의 회귀 계수와 절편 출력
print("회귀 계수 (기울기):", model.coef_)
print("절편 (y 절편):", model.intercept_)

# 예측
X_new = np.array([6, 7, 8]).reshape(-1, 1)  # 예측할 독립 변수 X
y_pred = model.predict(X_new)                # 예측 결과
print("예측 결과:", y_pred)
```
