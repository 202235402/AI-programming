```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
from sklearn.datasets import load_boston 
```

```python
boston = load_boston()
```

```python
df['PRICE'] = boston.target
```

```python
fig, axs = plt.subplots(figsize=(16,8), ncols=4, nrows=2)

lm_features = ['RM','ZN','INDUS','NOX','AGE','PTRATIO','LSTAT','RAD']

for i, feature in enumerate(lm_features):
    row = int(i/4) 
    col = i%4


    sns.regplot(x=feature, y='PRICE', data=df, ax=axs[row][col])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.model_selection import train_test_split

x_data=df.loc[:, ['LSTAT', 'RM']]
y_data=df.loc[:, 'PRICE']
X_train, X_test, y_train, y_test=train_test_split(x_data,
                                                  y_data,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train.head()
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
print("íšŒê·€ê³„ìˆ˜(ê¸°ìš¸ê¸°):", np.round(lr.coef_, 1)) 
print("ìƒìˆ˜í•­(ì ˆíŽ¸):", np.round(lr.intercept_, 1)) 
```

```python
y_preds = lr.predict(X_test)
print(X_train.shape, X_test.shape)
```

```python
mse = mean_squared_error(y_test, y_preds)
rmse = np.sqrt(mse)
print(f'MSE : {mse:.3f}, RMSE: {rmse:.3f}')
print(f'Variance score : {r2_score(y_test, y_preds):.3f}')
```

```python
print("ì ˆíŽ¸ ê°’:", lr.intercept_) 

print("íšŒê·€ê³„ìˆ˜:", np.round(lr.coef_,1))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X_test['LSTAT'], y_test, label='y_test')
plt.scatter(X_test['LSTAT'], y_preds, c='r', label='y_pred')
#plt.plot(X_train, lr.predict(X_train), color='blue')
plt.show()    
```
